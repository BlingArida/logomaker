```python
# Import LogoMaker
from logomaker import load_alignment, make_logo, PROTEIN
%matplotlib inline
```

```python
! head data/Pfam_wwdomain.fa
```

```python
matrix = load_alignment(fasta_file='data/Pfam_wwdomain.fa', characters=PROTEIN)
matrix.head()
```

```python
make_logo(matrix=matrix, figsize=[10,2]);
```

```python
make_logo(matrix=matrix,
          occurance_threshold=.2,
          figsize=[10,2]);
```

```python
make_logo(matrix=matrix,
          occurance_threshold=.2,
          logo_type='probability',
          figsize=[10,2]);
```

```python
make_logo(matrix=matrix,
          occurance_threshold=.2,
          colors='chemistry',
          logo_type='probability',
          figsize=[10,2]);
```

```python
logo, ax, fig = make_logo(
    matrix=matrix,
    occurance_threshold=.2,
    colors='chemistry',
    logo_type='probability',
    max_stretched_character='E',
    width=.8,
    stack_order='small_on_top',
    axes_style='light_rails',
    use_transparency=True,
    font_family='fantasy',
    figsize=[10,2],
    title='WW domain (Pfam PF00397).',
    save_to_file='output/wwdomain.pdf');
```

```python
fig.savefig
```
