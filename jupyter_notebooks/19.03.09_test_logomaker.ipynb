```python
from logomaker import Glyph

fig, ax = plt.subplots(figsize=[8,3])
ax.set_xlim([-1,2])
ax.set_ylim([-1,1])
Glyph(c='Z', p=0,ax=ax,ceiling=1,floor=-1)
Glyph(c='G', p=1, width=.75, alpha=.5, font_family='Arial Rounded MT Bold',ceiling=1,floor=-1, 
      color='dodgerblue', flip=False, mirror=True)
```

```python
fig, ax = plt.subplots(figsize=[15,3])

# Style axes
s = 'LOGOMAKER'
ax.set_ylim([0,1])
ax.set_xlim([-.5,len(s)-.5])
#ax.axis('off')

# Create a list of glyphs
glyphs = []
for p, c in enumerate(s):
    g = Glyph(c=c, p=p,ax=ax, draw_now=False, ceiling=1,floor=0,color='black')
    glyphs.append(g)
    
# Modify a single glyph
g = glyphs[5]
g.color='dodgerblue'
g.floor=.5
g.font_family='Comic Sans MS'
g.flip=False

# Draw all glyphs
for g in glyphs:
    g.draw()
    
```
