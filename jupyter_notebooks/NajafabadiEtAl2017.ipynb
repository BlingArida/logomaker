```python
from __future__ import division
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib as mpl
import re

# Import LogoMaker
import sys
sys.path.append('../../')
import logomaker

%matplotlib inline
plt.ion()
```

```python
logomaker.get_fontnames()
```

```python
counts_mat = pd.read_csv('../data/DBF.logo.txt',delim_whitespace=True)
```

```python
from matplotlib.ticker import FormatStrFormatter
import warnings
warnings.filterwarnings("ignore")


character_style_dict = {  'character_colors':{'DEC': [.42, .16, .42],
                                            'PG': [.47, .47, 0.0],
                                            'MIWALFV': [.13, .35, .61],
                                            'NTSQ': [.25, .73, .28],
                                            'RK': [.74, .18, .12],
                                            'HY': [.09, .47, .46]}
                        }

fullheight_style_dict = {
                         'fullheight_zorder':-10,
                         'fullheight_boxcolors':[.96, .95, .60],
                         'fullheight_vsep':0,
                         'fullheight_width':1,
                        }

baseline_param_dict = {
                          'baseline_style':':',
                          'baseline_width':1.5,
                      }

font_style_dict = {
                      'font_family':'Arial',
                      'font_weight':'heavy',
                  }


logo =logomaker.make_logo(dataframe=counts_mat,
                    figsize=[10,6],
                    bottom_spine=False,
                    character_style_dict=character_style_dict,
                    fullheight_style_dict=fullheight_style_dict,
                    baseline_param_dict=baseline_param_dict,      
                    vsep=.02,
                    draw_now=False,
                    max_stretched_character='M',
                    below_alpha=.5, 
                    remove_flattened_characters=True,
                    font_style_dict=font_style_dict,

                   )

fig, ax = plt.subplots(figsize=[12,7])

mpl.rcParams['font.size'] = 30
ax.xaxis.set_major_formatter(FormatStrFormatter('%+d'))

ax.tick_params
ax.set_ylabel('Alpha helix poxition')
ax.set_xlabel('Contribution to DBF model')

logo.draw(ax)

fig.savefig('../output/NajafabadiEtAl2017.pdf')
```
