```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
plt.ion()

import sys
sys.path.append('../../')

import logomaker
```

```python
# Create a logo
df = pd.read_csv('../data/rnap_tau_final_all.41.matrix.txt', delim_whitespace=True)
df.head()
```

```python
#fig, ax = plt.subplots(figsize=[12,3])

# Create a Logo
logo = logomaker.Logo(df)
```

```python
#fig, ax = plt.subplots(figsize=[12,3])

# Using data frame, create matrix object
#mat = logomaker.Matrix(df)
#mat = logomaker.validate_matrix(df)

# Create a BaseLogo
logo = logomaker.Logo(df=df,
                      negate=True,
                      center=True,
                      draw_now=False)

# Style glyphs
logo.style_glyphs(alpha=.9,
                  color='gray',
                  font_family='Arial Rounded MT Bold')

# Add below effects
#logo.add_below_effects(shade=.3, fade=.5)

# Modify a single glyph
logo.style_single_glyph(p=30, c='A', color='dodgerblue')

# Highlight wt sequence
wt_seq = 'CAGGCTTTACACTTTATGCTTCCGGCTCGTATGTTGTGTGG'
logo.style_glyphs_in_sequence(wt_seq, 
                              edgewidth=1.0, 
                              alpha=1.0, 
                              color='yellow')

# Draw a baseline
logo.draw_baseline(linewidth=.5)

# Style xticks
logo.style_xticks(spacing=5, anchor=25, rotation=90, fmt='%d')

# Style spines
logo.style_spines(visible=False)
logo.style_spines(visible=True, spines=['top', 'bottom'], linewidth=2)

# Draw logo
logo.draw()
```
