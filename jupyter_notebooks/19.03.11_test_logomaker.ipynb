```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
plt.ion()

import sys
#sys.path.append('/Users/jkinney/github/')
sys.path.append('../../')

import logomaker
```

```python
# Create a logo
df = pd.read_csv('../data/rnap_tau_final_all.41.matrix.txt', delim_whitespace=True)
df.head()
```

```python
from logomaker.BasicLogo import BasicLogo

fig, ax = plt.subplots(figsize=[12,3])

# Create a BaseLogo
logo = BasicLogo(df)
```

```python
fig, ax = plt.subplots(figsize=[12,3])

from logomaker.Matrix import Matrix

# Using data frame, create matrix object
mat = Matrix(df)

# Create a BaseLogo
logo = BasicLogo(matrix=mat,
                 negate=True,
                 center=True,
                 draw_now=False)

# Style glyphs
logo.style_glyphs(alpha=.9,
                  color='gray',
                  font_family='Impact', 
                  dont_stretch_more_than='A',
                  width=.9,
                  vpad=.1)

# Add below effects
logo.add_below_effects(shade=.3, fade=.5)

# Modify a single glyph
logo.style_single_glyph(p=30, c='A', color='dodgerblue')

# Highlight wt sequence
wt_seq = 'CAGGCTTTACACTTTATGCTTCCGGCTCGTATGTTGTGTGG'
logo.style_glyphs_in_sequence(wt_seq, 
                              edgewidth=1.0, 
                              alpha=1.0, 
                              color='yellow')

# Draw logo
logo.draw()
```
