```python
from __future__ import division
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import pandas as pd

# Import LogoMaker
import logomaker

%matplotlib inline
plt.ion()
```

```python
# Load counts matrix from fasta file
counts_mat = logomaker.load_alignment(fasta_file='data/crp_sites.fasta')
```

```python
# Make all logo types
wtseq = 'ATTAATGTGAGTTAGCTCACTCATTA'

parameters_txt = """
logo_type : 'probability'
below_flip : True
foo bar
virus : eval(x=5)
highlight_edgewidth : 1
colors : 'classic'
highlight_colors: 'classic'
highlight_sequence : 'ATTAATGTGAGTTAGCTCACTCATTA'
vpad : .05
width : .90
"""
with open('parameters.txt','w') as f:
    f.write(parameters_txt)

logo = logomaker.make_styled_logo(style_file='parameters.txt', matrix=counts_mat)

fig, ax = plt.subplots(figsize=[6,1.5])

logo.draw(ax)

fig.tight_layout()
```
