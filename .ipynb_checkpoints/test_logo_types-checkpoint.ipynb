```python
from __future__ import division
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import pandas as pd

# Import LogoMaker
import logomaker

%matplotlib inline
plt.ion()
```

```python
# Load counts matrix from fasta file
counts_mat = logomaker.load_alignment(file_name='crp_sites.fasta')
```

```python
# Make all logo types
wtseq = 'ATTAATGTGAGTTAGCTCACTCATTA'

parameters_txt = """
logo_type : 'enrichment'
colors : 'random'
hpad : .5
vpad : 0.1
below_flip : False
"""
with open('parameters.txt','w') as f:
    f.write(parameters_txt)

logo1 = logomaker.make_logo(counts_mat, colors='classic')
logo2 = logomaker.make_styled_logo('parameters.txt', counts_mat)

fig, axs = plt.subplots(2, figsize=[8,4])

logo1.draw(axs[0])
logo2.draw(axs[1])

fig.tight_layout()
```
