```python
from __future__ import division
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import pandas as pd

# Import LogoMaker
import logomaker

%matplotlib inline
plt.ion()
```

```python
# Load counts matrix from fasta file
counts_mat = logomaker.load_alignment(file_name='crp_sites.fasta')
```

```python
# Make all logo types
wtseq = 'ATTAATGTGAGTTAGCTCACTCATTA'

parameters_txt = """
logo_type : 'probability'
colors : 'random'
ylabel : 'Probability'
hpad : .5
vpad : 0.1
foo : 'bar'
"""
with open('parameters.txt','w') as f:
    f.write(parameters_txt)

probability_logo = logomaker.make_logo(counts_mat, parameters_file='parameters.txt', bar='baz')
fig, ax = plt.subplots(figsize=[8,2])
probability_logo.draw(ax)

fig.tight_layout()
```
