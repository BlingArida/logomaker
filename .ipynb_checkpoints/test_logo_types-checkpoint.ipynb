```python
from __future__ import division
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import pandas as pd

# Import LogoMaker
import logomaker

%matplotlib inline
plt.ion()
```

```python
# Load counts matrix from fasta file
counts_mat = logomaker.load_alignment('crp_sites.fasta')
```

```python
# Make all logo types
wtseq = 'ATTAATGTGAGTTAGCTCACTCATTA'

counts_logo = logomaker.make_logo(counts_mat, logo_type='counts', 
                                  use_transparency=False,
                                  alpha=.5)
probability_logo = logomaker.make_logo(counts_mat, logo_type='probability', 
                                       colors='cool',
                                       boxcolors='black',
                                       boxalpha=1,
                                       highlight_sequence=wtseq,
                                       highlight_boxcolors='gray',
                                       highlight_boxalpha=.5)
enrichment_logo = logomaker.make_logo(counts_mat, logo_type='enrichment')
energy_logo = logomaker.make_logo(counts_mat, logo_type='energy')
information_logo = logomaker.make_logo(counts_mat, logo_type='information')

# Draw logos
fig, ax_list = plt.subplots(5, figsize=[8,10])
counts_logo.draw(ax_list[0])
probability_logo.draw(ax_list[1])
enrichment_logo.draw(ax_list[2])
energy_logo.draw(ax_list[3])
information_logo.draw(ax_list[4])

fig.tight_layout()
```
