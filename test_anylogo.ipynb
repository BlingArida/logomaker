```python
from __future__ import division
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import pandas as pd

# Logo-generating module
import anylogo

%matplotlib inline
plt.ion()
```

```python
# Load energy matrix
energy_mat = pd.read_csv('crp_fullwt.26.txt',delim_whitespace=True)

mat = anylogo.data.validate_mat(energy_mat)
freq_mat = anylogo.data.transform_mat(energy_mat, from_type='energy_mat', to_type='freq_mat')
freq_mat.index.name='pos'

freq_mat.to_csv('crp_freq_mat.txt', sep='\t', float_format='%.2f')
freq_mat.head()
```

```python
fig = plt.figure(figsize=[3.3,3.5])

import matplotlib as mpl
mpl.rcParams['lines.linewidth'] = .5
mpl.rcParams['font.size'] = 5
mpl.rcParams['axes.labelsize'] = 8

# Plot frequency logo
ax = fig.add_subplot(3,1,1)
anylogo.Logo(mat=energy_mat, mat_type='energy_mat', font_name='Arial Bold',
             logo_type='freq_logo', color_scheme='black', logo_style='rails', 
             stack_order='small_on_top', use_transparency=True).draw()
plt.title('probability logo', fontsize=8)

# Plot information logo
ax = fig.add_subplot(3,1,2)
anylogo.Logo(mat=energy_mat, mat_type='energy_mat', 
             logo_type='info_logo', color_scheme='gist_rainbow').draw()
plt.title('information logo', fontsize=8)

# Plot energy logo
ax = fig.add_subplot(3,1,3)
anylogo.Logo(mat=energy_mat, mat_type='energy_mat', 
             logo_type='energy_logo', neg_flip=True, 
             logo_style='everything', font_name='Comic Sans MS Bold', color_scheme='classic').draw()
plt.title('energy logo', fontsize=8)

fig.tight_layout(h_pad=2)
fig.savefig('logos.pdf')
```
