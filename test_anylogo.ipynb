```python
from __future__ import division
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import pandas as pd

# Logo-generating module
import anylogo

%matplotlib inline
plt.ion()
```

```python
# Load energy model and tidy up
effect_df = -pd.read_csv('crp.txt',delim_whitespace=True)
effect_df.head()
```

```python
fig = plt.figure(figsize=[8,6])

# Probability logo
ax = fig.add_subplot(3,1,1)
anylogo.draw(ax,effect_df=effect_df,logo_type='probability', logo_style='everything')

# Information logo
ax = fig.add_subplot(3,1,2)
anylogo.draw(ax,effect_df=effect_df,logo_type='information')

# Effect logo
ax = fig.add_subplot(3,1,3)
anylogo.draw(ax,effect_df=effect_df,logo_type='effect')

# Save logos
plt.tight_layout()
plt.savefig('logos.pdf')
```
