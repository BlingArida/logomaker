```python
from __future__ import division
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import pandas as pd

# Import LogoMaker
import logomaker

%matplotlib inline
plt.ion()
```

```python
# Load counts matrix from fasta file
counts_mat = logomaker.load_alignment('crp_sites.fasta')
counts_mat.to_csv('crp_counts_mat.txt', sep='\t', float_format='%d')
```

```python
import matplotlib as mpl
mpl.rcParams['lines.linewidth'] = .5
mpl.rcParams['font.size'] = 5
mpl.rcParams['axes.labelsize'] = 8

# Make information logo
# THIS INFORMATION LOGO IS NOT RENDERING PROPERLY
logo1 = logomaker.Logo(mat=counts_mat, font_weight='bold',
             logo_type='info_logo', color_scheme='gist_rainbow', 
             logo_style='classic', ylim=[0,1])

# Make frequency logo
logo2 = logomaker.Logo(mat=counts_mat, font_family='Impact',
             logo_type='freq_logo', color_scheme='blue', logo_style='rails', 
             stack_order='small_on_top', use_transparency=True)

# Make energy logo
logo3 = logomaker.Logo(mat=counts_mat, font_family='Comic Sans MS',
             font_weight='bold',
             logo_type='energy_logo', neg_flip=True, 
             logo_style='everything',
             color_scheme='classic')

# Draw logos
fig, ax_list = plt.subplots(3, figsize=[3.3,2.8])
logo1.draw(ax_list[0])
logo2.draw(ax_list[1])
logo3.draw(ax_list[2])

fig.tight_layout(h_pad=2)
fig.savefig('logos.pdf')
```
