```python
from __future__ import division
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib as mpl
import re

# Import LogoMaker
import sys
sys.path.append('../../')
from logomaker import load_alignment, make_logo
%matplotlib inline
```

```python
import warnings
warnings.filterwarnings("ignore")
matrix = load_alignment(fasta_file='../data/Pfam_wwdomain.fa', sequence_type='PROTEIN')
```

```python
mpl.rcParams['font.size'] = 20

logo = make_logo(dataframe=matrix,
          character_style_dict={'character_colors':'hydrophobicity'},   
          draw_now=False,
         )

fig, ax = plt.subplots(figsize=[12,3])

logo.draw(ax)
fig.savefig('../output/Pfam_wwdomain.pdf')
```

```python
logo = make_logo(dataframe=matrix,
          character_style_dict={'character_colors':'blue'},   
          draw_now=False,
          matrix_type='probability'
         )

fig, ax = plt.subplots(figsize=[12,3])
ax.set_ylabel('probability')
logo.draw(ax)
```

```python
font_style_dict = {
                      'font_family':'fantasy',
                      'font_weight':'heavy',
                  }


logo = make_logo(
    dataframe=matrix,
    character_style_dict={'character_colors':'charge'},   
    font_style_dict = font_style_dict,
    matrix_type='probability',
    max_stretched_character='E',
    stack_order='small_on_top',
    axes_type='light_rails',
    draw_now=False,
    use_transparency=True,
)

fig, ax = plt.subplots(figsize=[12,3])
ax.set_ylabel('probability')
ax.set_title('WW domain (Pfam PF00397)')
logo.draw(ax)
fig.savefig('../output/wwdomain.pdf')
```
