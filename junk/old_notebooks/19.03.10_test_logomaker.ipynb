```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
plt.ion()

import sys
#sys.path.append('/Users/jkinney/github/')
sys.path.append('../../')

import logomaker
```

```python
# Create a logo
df = pd.read_csv('../data/rnap_tau_final_all.41.matrix.txt', delim_whitespace=True)
df.head()
```

```python
from logomaker.BasicLogo import BasicLogo

fig, ax = plt.subplots(figsize=[12,3])

# Create a BaseLogo
logo = BasicLogo(ax=ax, matrix=df)

logo.draw()
```

```python
fig, ax = plt.subplots(figsize=[12,3])

# Create a BaseLogo
logo = BasicLogo(ax=ax, matrix=df,
                negate=True,
                center=True,
                colors='salmon',
                alpha=.9,
                vpad=.1, 
                dont_stretch_more_than='A',
                font_family='Arial Rounded MT Bold',
                shade_below=.5, 
                fade_below=.1,
                width=0.9,
                draw_now=False)

# Highlight wt sequence
wt_seq = 'CAGGCTTTACACTTTATGCTTCCGGCTCGTATGTTGTGTGG'
highlight_dict = {
    'edgewidth':1.0,
    'alpha':1.0,
    'color':'salmon'
}
logo.highlight_sequence(wt_seq, highlight_dict)

# Modify a single glyph
logo.modify_glyph(p=30, c='A', params_dict={'color':'dodgerblue'})

# Draw logo
logo.draw()
```
