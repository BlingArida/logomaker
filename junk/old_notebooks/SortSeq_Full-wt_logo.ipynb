```python
from __future__ import division
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import pandas as pd

# Import LogoMaker
import sys
sys.path.append('../../')
import logomaker

%matplotlib inline
plt.ion()
```

```python
# Load counts matrix from files
df_wt = pd.read_csv('../data/full-wt-formatted.txt', delim_whitespace=True)
counts_wt_mat = logomaker.load_alignment(sequences=df_wt['seq'], sequence_counts=df_wt['ct_9'])
bg_mat = logomaker.load_alignment(sequences=df_wt['seq'], sequence_counts=df_wt['ct_0'])
enrichment_mat = np.log2(counts_wt_mat/bg_mat)
```

```python
# Make logos

highlight_style_param_dict = {
                              'highlight_sequence':'AATTAATGTGAGTTAGCTCACTCATTAGGCACCCCAGGCTTTACACTTTATGCTTCCGGCTCGTATGTTGTGTGG',   
                              'highlight_colors':'tomato'
                               }
scalebar_param_dict = {'scalebar_text':'2-fold'}

logomaker.make_logo(dataframe=enrichment_mat,
                    center_columns=True, 
                    character_style_dict = {'character_colors':'gray'}, 
                    highlight_style_dict = highlight_style_param_dict,
                    scalebar_dict=scalebar_param_dict,
                    axes_type='scalebar', 
                    figsize=[14,4],
                   )

plt.savefig('../output/plac_enrichment.pdf')
```
